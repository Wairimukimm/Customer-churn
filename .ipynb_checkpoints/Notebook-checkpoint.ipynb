**1. BASELINE MODEL -- DECISION TREE CLASSIFIER**

```python
class CustomDecisionTree:
    def __init__(self, custom_parameter=1):
        self.custom_parameter = custom_parameter
        self.model = DecisionTreeClassifier(random_state=42)
        self.selector = None
        
    def fit(self, X_train_scaled, y_train):
        #feature selection
        self.selector = SelectFromModel(self.model, threshold='median')
        self.selector.fit(X_train_scaled, y_train)

        # Transform the training data to include selected features
        X_train_selected = self.selector.transform(X_train_scaled)

        # Fit the model on the selected features
        self.model.fit(X_train_selected, y_train)

    def predict(self, X_test_scaled):
        X_test_selected = self.selector.transform(X_test_scaled)
        return self.model.predict(X_test_selected)
    
    def plot_feature_importance(self, feature_names):
        importances = self.model.feature_importances_
        indices = np.argsort(importances)[::-1]

        plt.figure(figsize=(10, 6))
        plt.bar(range(len(indices)), importances[indices], align='center')
        plt.xticks(range(len(indices)), [feature_names[i] for i in indices], rotation=90)
        plt.xlabel('Feature')
        plt.ylabel('Importance')
        plt.title('Feature Importances')
        plt.tight_layout()
        plt.show()

    def plot_confusion_matrix(self, X_test_scaled, y_test):
        X_test_selected = self.selector.transform(X_test_scaled)
        predictions = self.model.predict(X_test_selected)
        cm = confusion_matrix(y_test, predictions, normalize='true')
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

    def plot_roc_curve(self, X_test_scaled, y_test, label=None):
        X_test_selected = self.selector.transform(X_test_scaled)
        y_scores = self.model.predict_proba(X_test_selected)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_scores)
        roc_auc = auc(fpr, tpr)

        plt.plot(fpr, tpr, label=f'{label} (AUC = {roc_auc:.2f})')
```

```python
# Instantiate CustomDecisionTree
clf = CustomDecisionTree()

# Fit the model and make predictions
clf.fit(X_train_scaled, y_train_resampled)
pred = clf.predict(X_test_scaled)

# Calculate recall score and accuracy score
recall = recall_score(y_test, pred)
accuracy = accuracy_score(y_test, pred)

# Print recall score and accuracy
print(f"Recall Score: {recall}")
print(f"Accuracy of the model: {accuracy}")

# Plot confusion matrix
clf.plot_confusion_matrix(X_test_scaled, y_test)
```

RATIONALE
* I used decision trees as a basline model because it has the ability to capture non linear relationships within the data and it simple to interpret.

RESULTS
* Recall: Recall calculates the proportion of correctly identified churned customers among all actual churned customers. This is particularly critical for a bank, as failing to identify actual churned customers could result in significant revenue loss.

* The recall for the decision tree model is 67%. A recall score of 0.67 indicates that the model correctly identified approximately 67% of the total positive instances (churned customers) in the dataset.

* Overall model accuracy is 74%, indicating that the model correctly predicted the churn or no-churn status for approximately 74% of all customers.

LIMITATIONS
* Small variations in the training data can result in completely different trees. This instability can affect the reliability of the model.
* Approximately 32% of the customers who churned were not identified by the model (False Negatives). 
  False Negatives (FN): These are instances where the model predicted customers as non-churned, but they actually churned.
  
RECOMMENDATIONS
* Consider using random forest because of Improved Accuracy: Random Forests typically provide higher accuracy compared to individual Decision Trees, especially in complex tasks. They are less sensitive to noise and outliers in the data due to the averaging effect across multiple trees.

**2. RANDOM FOREST CLASSIFIER**

```python
class CustomRandomForest(CustomDecisionTree):
    def __init__(self, n_estimators=100, max_depth=None, min_samples_split=3):
        super().__init__()
        self.model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_samples_split)

    def perform_cross_validation(self, X, y, cv=5, scoring='recall'):
        cv_scores_recall = cross_val_score(self.model, X, y, cv=cv, scoring=scoring)
        cv_scores_accuracy = cross_val_score(self.model, X, y, cv=cv, scoring='accuracy')
        print(f"Cross-validation {scoring.capitalize()} Scores:", cv_scores_recall)
        print(f"Mean {scoring.capitalize()} Score:", cv_scores_recall.mean())
        print(f"Cross-validation Accuracy Scores:", cv_scores_accuracy)
        print(f"Mean Accuracy Score:", cv_scores_accuracy.mean())

    def fit_and_cross_validate(self, X, y, cv=5, scoring='recall'):
        self.fit(X, y)
        self.perform_cross_validation(X, y, cv=cv, scoring=scoring)

# instantaite the random forest
rf = CustomRandomForest(n_estimators=100, max_depth=5, min_samples_split=5)

# Fit and cross-validate
rf.fit_and_cross_validate(X_train_scaled, y_train_resampled)

# Plot confusion matrix
rf.plot_confusion_matrix(X_test_scaled, y_test)

# Make predictions
pred = rf.predict(X_test_scaled)

# Calculate recall score and accuracy score
recall = recall_score(y_test, pred)
accuracy = accuracy_score(y_test, pred)

# Print scores
print(f"Recall Score : {recall}")
print(f"Accuracy of the model: {accuracy}")
```

RESULTS FOR RANDOM FOREST

RATIONALE
* Decision trees tend to overfit the training data. Random Forests mitigate this issue by aggregating multiple trees. This helps in tackling overfitting and increasing generalization to unseen data.

* Improved Accuracy: Random Forests often provide higher accuracy than a single decision tree due to their ensemble nature. They leverage the wisdom of crowds by combining predictions from multiple trees.

RESULTS
* Recall: Recall calculates the proportion of correctly identified churned customers among all actual churned customers. This is particularly critical for a bank, as failing to identify actual churned customers could result in significant revenue loss.

* The recall for the decision tree model is 69%. A recall score of 0.69 indicates that the model correctly identified approximately 69% of the total positive instances (churned customers) in the dataset

* Accuracy: Overall model accuracy is 80%, indicating that the model correctly predicted the churn or no-churn status for approximately 80% of all customers.

LIMITATIONS
* It is hard to interpret random forest. Understanding the overall decision-making process might be challenging due to the ensemble nature of the model.

HYPERPARAMETER TUNING FOR RANDOM FOREST

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 3, 5]
}

# Create a GridSearchCV and fit
grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring='recall')
grid_search.fit(X_train_scaled, y_train_resampled)

# Get the best parameters and the best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Use the best estimator found by GridSearchCV
best_rf = grid_search.best_estimator_

# Make predictions
pred = best_rf.predict(X_test_scaled)

# Calculate recall score and accuracy score
recall = recall_score(y_test, pred)
accuracy = accuracy_score(y_test, pred)

# Print best parameters and scores
print("Best Parameters:", best_params)
print("Best Recall Score:", best_score)
print(f"Recall Score: {recall}")
print(f"Accuracy Score: {accuracy}")
```

The random forest doesn't improve even after doing gridsearch to find the best hyperparameters.

**3. XG BOOST**

```python
class CustomGradientBoosting(CustomDecisionTree):
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=3):
        super().__init__() 
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.model = XGBClassifier(
            n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth)

xgb_boost = CustomGradientBoosting(n_estimators=100, learning_rate=0.1)
xgb_boost.fit(X_train_scaled, y_train_resampled)

# Make predictions
pred = xgb_boost.predict(X_test_scaled)

# Calculate recall score and accuracy score
recall = recall_score(y_test, pred)
accuracy = accuracy_score(y_test, pred)

# Print the recall score and accuracy
print(f"Recall Score : {recall}")
print(f"Accuracy of the model: {accuracy}")

# Plot confusion matrix 
xgb_boost.plot_confusion_matrix(X_test_scaled, y_test)
```

RESULTS FOR XGBOOST

RATIONALE

XGBoost has techniques like regularization and a robust objective function that can handle class imbalance better than Random Forest. In customer churn scenarios, where the positive class (churned customers) might be significantly smaller than the negative class (non-churned), XGBoost might provide better performance.

RESULTS

Recall: Recall calculates the proportion of correctly identified churned customers among all actual churned customers. This is particularly critical for a bank, as failing to identify actual churned customers could result in significant revenue loss.

The recall for the decision tree model is 70%. A recall score of 0.70 indicates that the model correctly identified approximately 70% of the total positive instances (churned customers) in the dataset

**4. GRADIENT BOOSTING**

```python
# Inherits attributes from CustomDecisionTree
class CustomGradientBoosting(CustomDecisionTree):
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=3):
        super().__init__() 
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.model = GradientBoostingClassifier(
            n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth)
gradient_boost = CustomGradientBoosting(n_estimators=100, learning_rate=0.1)
gradient_boost.fit(X_train_scaled, y_train_resampled)

#make predictions
pred = gradient_boost.predict(X_test_scaled)

# Calculate recall score and accuracy score
recall = recall_score(y_test, pred)
accuracy = accuracy_score(y_test, pred)

# Print the recall score
print(f"Recall Score : {recall}")
print(f"Accuracy of the model: {accuracy}")

#Plot confusion matrix 
gradient_boost.plot_confusion_matrix(X_test_scaled, y_test)
```

complex relationships within our data well.
